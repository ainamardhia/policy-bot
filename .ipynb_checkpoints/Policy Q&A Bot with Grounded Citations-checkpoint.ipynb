# Project - Policy Q&A Bot with Grounded Citations

### Scenario:
Build a Q&A assistant over a small set of **policy wording documents** (PDF/DOCX/Markdown). Users ask questions like:
- "Is wear-and-tear covered under the standard policy?"
- "What is the waiting period for accidental damage?"
- "How does the deductible apply to water damage?"
- "What definitions apply to 'Insured Person'?"

The assistant must answer **strictly from the policy wordings** and include **clause-level citations** (doc name + section/heading + clause number or page).

### Core Requirements
- Implement RAG pipiline in Python:
  - **Ingestion**: Load 2-3 policy wording docs. Support PDF at minimum.
  - **Chunking**: Use **section-aware chunking**:
    - Preserve **headings** (e.g., *Definitions*, *Exclusions*, *Coverage*, *Conditions*, *Endorsements*).
    - Capture **clause numbers** and **cross-references** ("see Section 3.2").
    - Chunk size: 400-800 tokens with overlap; attach metadata: `{doc_name, section, clause_number, page, heading_path)`.
  - **Vector Store**: FAISS or Chroma.
  - **Retriever -> LLM -> Answer**: Retrieve top-k relevant chunks, construct context, amd generate grounded answers
- **Citations (Mandatory)**
  - Each answer must end with citations like:
    `Sources: Policy_A.pdf ยง3.2 (Exclusions), p.12; Policy_B.docx "Definitions > Accident", p.4`
